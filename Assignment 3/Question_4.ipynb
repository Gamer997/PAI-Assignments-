```python
import pandas as pd
import openpyxl
# Specify the Excel file
excel_file = "timetable.xlsx"

# Read all sheets into a dictionary
all_sheets = pd.read_excel(excel_file, header=None, skiprows=4, sheet_name=None)

# Loop through the sheets and process them as needed
for sheet_name, df in all_sheets.items():
    # Drop columns with no data
    df = df.dropna(axis=1, how='all')

    # Extract the row with the element location (assuming it's always row 2)
    #location_of_element = df.loc[2]

    # Print the sheet name
    print(f"Sheet Name: {sheet_name}")

    # Print the row with the element location
    #print(location_of_element)

    # Print the DataFrame
    #print(df)

# Load the Excel file
file_path = 'timetable.xlsx'  # Replace with the path to your Excel file
workbook = openpyxl.load_workbook(file_path)

# Create a new workbook to store the results
new_workbook = openpyxl.Workbook()

# Get user input
search_value = input(f"Enter the value to search for in the '{sheet_name}' sheet: ")


# Iterate through all sheets in the input workbook
for sheet_name in workbook.sheetnames:
    # Select the current sheet
    sheet = workbook[sheet_name]


    # Flag to keep track if the value is found
    value_found = False

    # Initialize rgb_color variable
    rgb_color = None

    # for finding the rgb color value of any cell
    for row in sheet.iter_rows(min_row=1, max_row=40, min_col=1, max_col=sheet.max_column):
        for cell in row:
            if cell.value == search_value:
                # Get the RGB value of the cell's fill color
                if cell.fill.start_color is not None:
                    rgb_color = cell.fill.start_color.rgb
                    value_found = True
                    break

    # If the value is not found
    if not value_found:
        print(f'The value "{search_value}" was not found in the first four rows of the "{sheet_name}" sheet.')
    else:
        # Create a new sheet in the new workbook with the same name as the current sheet
        new_sheet = new_workbook.create_sheet(title=sheet_name)

        # for finding those cells which have the same rgb color value
        if rgb_color:
            for row in sheet.iter_rows(min_row=4, max_row=sheet.max_row, min_col=1, max_col=sheet.max_column):
                for cell in row:
                    if cell.fill.start_color.rgb == rgb_color:
                        cell_coordinate = cell.coordinate
                        cell_value = cell.value  # Get the cell's value

                        # Set the value in the new sheet using the cell coordinate
                        new_sheet[cell_coordinate] = cell_value

        # Copy values from specific ranges in the original sheet to the new sheet
        for row_index in range(1, 42):
            original_value = sheet.cell(row=row_index, column=1).value
            new_sheet.cell(row=row_index, column=1, value=original_value)

        for col_index in range(1, 24):
            original_value = sheet.cell(row=5, column=col_index).value
            new_sheet.cell(row=5, column=col_index, value=original_value)

# Remove the default sheet created by openpyxl in the new workbook
new_workbook.remove(new_workbook['Sheet'])

# Save the new workbook
new_workbook.save('result.xlsx')
```
