```python
import pandas as pd


course_allocator_df = pd.read_excel('course_instructors.xlsx', skiprows=2)

course_allocator_df['Course'].fillna(method='ffill', inplace=True)
course_allocator_df['Section'].fillna(method='ffill', inplace=True)

# Create a dictionary to map teachers to their assigned classes
teacher_to_classes = {}
current_teacher = None  # Initialize the current_teacher variable
previous_course_name = None  # Initialize the previous_course_name variable

for index, row in course_allocator_df.iterrows():
    course_name = row['Course']  # Update with the correct column name
    section = row['Section']  # Update with the correct column name
    teacher_name = row['Course Instructor']  # Update with the correct column name


    if not pd.isna(course_name) and course_name.strip().lower() != 'naan':
        current_teacher = teacher_name
        previous_course_name = course_name  # Update the previous_course_name

    # Add the course to the current teacher

    if current_teacher:
        if current_teacher not in teacher_to_classes:
            teacher_to_classes[current_teacher] = [(course_name, section)]
        else:
            teacher_to_classes[current_teacher].append((course_name, section))

# Function to get classes assigned to a teacher
def get_classes_for_teacher(teacher_name):
    if teacher_name in teacher_to_classes:
        return teacher_to_classes[teacher_name]
    else:
        return []

# Example: Get classes assigned to a specific teacher (User Input)
user_input_teacher = input("Enter the name of the teacher: ")
assigned_classes = get_classes_for_teacher(user_input_teacher)
if assigned_classes:
    print(f"Classes assigned to {user_input_teacher}:")
    for course, section in assigned_classes:
        print(f"Course: {course}, Section: {section}")
else:
    print(f"{user_input_teacher} is not assigned to any classes.")

timetable_df = pd.read_excel('timetable.xlsx', header=None, skiprows=4, sheet_name = "Thursday")

excel_file = "timetable.xlsx"  

df = pd.read_excel(excel_file, header=None, skiprows=4)
df = df.dropna(axis=1, how='all')


```
