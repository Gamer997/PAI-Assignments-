```python
import pandas as pd

# Read the timetable data from an Excel file
timetable_data = pd.read_excel('TimeTable, FSC, Fall-2023.xlsx')

# Define the batch you want to extract
target_batch = "CS 2022"  # Change this to your desired batch

# Create an empty DataFrame to store the timetable for the target batch
timetable_df = pd.DataFrame(columns=["Course", "CHs", "Section", "Course Instructor", "Course Coordinator", "Time", "Room", "Day"])

# Iterate through the rows of the timetable data to extract information for the target batch
for index, row in timetable_data.iterrows():
    section = row["Section"]
    if target_batch in section:
        course_code = row["Code"]
        course_name = row["Course"]
        instructor = row["Course Instructor"]
        coordinator = row["Course Coordinator"]
        # Assuming that the time, room, and day information are stored in the "Time", "Room", and "Day" columns of the DataFrame.
        time = row["Time"]
        room = row["Room"]
        day = row["Day"]
        
        # Add the extracted information to the DataFrame
        timetable_df = timetable_df.append({
            "Course": course_code + " - " + course_name,
            "CHs": row["CHs"],
            "Section": section,
            "Course Instructor": instructor,
            "Course Coordinator": coordinator,
            "Time": time,
            "Room": room,
            "Day": day
        }, ignore_index=True)

# Export the timetable for the target batch to an Excel file
timetable_df.to_excel("CS_2022_TimeTable.xlsx", index=False)
```
