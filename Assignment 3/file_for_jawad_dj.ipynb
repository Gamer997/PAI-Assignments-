```python
import pandas as pd

# Load the Excel file
excel_file = "timetable.xlsx"
xl = pd.ExcelFile(excel_file)

# Initialize an empty list to store DataFrames for each sheet
data_frames = []

# Loop through the sheets in the Excel file
for sheet_name in xl.sheet_names:
    # Read each sheet into a DataFrame
    df = xl.parse(sheet_name)
    
    # Add a new column with the sheet name to identify the source of data
    df['Sheet Name'] = sheet_name
    
    # Append the DataFrame to the list
    data_frames.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(data_frames, ignore_index=True)

# Save the combined data to a CSV file
csv_file = "combined_timetable.csv"
combined_df.to_csv(csv_file, index=False)

print(f"Data from all sheets has been combined and saved to {csv_file}.")
```
